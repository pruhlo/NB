### Створення локального репозиторію, пов'язаного з віддаленим репозиторієм

Наступним кроком після створення репозиторію на github, званого далі віддаленим репозиторієм, є створення локальної копії цього репозиторію на своєму комп'ютері. **Особливістю** git **є наявність** на локальному комп'ютері **повної копії** репозиторію з усією інформацією про історію змін.

> **Відкриваємо консольний клієнт**.<br>
На Windows після інсталяції клієнта з'являється пункт Git Bash у контекстному меню папки. 
Достатньо перейти в бажану папку та скористатися цим пунктом меню.<br>
На Unix системах достатньо відкрити термінал та перейти у потрібну директорію. При стандартному встановленні консольного клієнта буде доступна команда git без додаткових зусиль.

> **Виконуємо команду** `git clone https://github.com/%user_login%/%repo_name%.git`<br> 
Повне https посилання на репозиторій для його викачування можна знайти на сторінці самого репозиторію на github. Після цього у поточній папці з'явиться нова папка з ім'ям %repo_name%, що містить копію віддаленого (remote) репозиторію.

>>У **випадку** коли потрібно підєднати якусь папку з даними до репозиторію спочатку потрібно ініціалізувати локальний репозиторій у відповідній директорії командою<br>
`git init`<br>
>>>для відміни цієї команди потрібно видалити .git директорію командою<br>
`rm -rf .git` 

>>потім **Виконайте команду**<br>
`git remote`

Оскільки пов'язаного репозиторію не існує, з Git не було отримано жодних вихідних даних.
Тепер за допомогою наведеної вище URL ми зв'яжемо репозиторій. Щоб зв'язати репозиторій, виконайте таку команду та натисніть клавішу enter:<br>
`remote add origin https://github.com/%user_login%/%repo_name%.git`<br>
Як тільки це буде зроблено, локальний репозиторій буде пов'язаний із репозиторієм GitHub<br>
Щоб перевірити, чи зв'язали ми наш репозиторій, знову виконайте команду `git remote` або команду `git remote-v` для перегляду того ж результату разом з URL-адресою


> **Переходимо в нову папку репозиторію і налаштовуємо його:**<br>
`git config user.name taras.shevchenko`<br>
`git config user.email taras.shevchenko@somemail.com`

### Внесення та оформлення змін у локальному репозиторії

> Скориставшись командою `git status` можна дізнатися, на якій гілці (branch) репозиторію ви зараз знаходитесь, які зміни є у вашій робочій копії та іншу інформацію. Робочою копією називається сукупність файлів у локальній папці репозиторію крім службових файлів.
 
<!-- https://www.lenakso.top/podklyuchenie-lokalnogo-repozitoriya-k-udalennomu-repozitoriyu-github/ -->

>Після внесення будь-яких змін до робочої копії їх можна "закомітити" в локальний репозиторій:
спочатку потрібна частина змін готується до коміту за допомогою команди<br> `git add %file_path%`
після чого проводиться коміт командою<br> `git commit`<br> Використання команди без аргументів відкриє текстовий редактор, де треба буде написати коментар для комміта, комміт обов'язково повинен мати коментар. Іншим варіантом завдання коментаря до комміту є використання команди<br> `git commit -m "%commit_message%"` 

>Історію змін можна переглянути командою<br> 
`git log` або `git log --name-only`<br>
Якщо вся історія змін не вміщується на екрані, можна користуватися клавішами прокручування на клавіатурі ("стрілочки", PgUp, PgDown), вихід з режиму перегляду змін здійснюється натисканням клавіші "q".

### Завантаження локальних змін у віддалений репозиторій

Після того, як були виконані необхідні локальні комміти, зміни можна завантажити у віддалений репозиторій за допомогою команди<br>
`git push origin master`<br>
GIT клієнт при цьому запросить ім'я користувача та пароль для доступу до github. Виконання цієї команди може закінчитися з помилкою, якщо в локальному репозиторії відсутні останні зміни, які є у віддаленому репозиторії. Для вирішення цієї проблеми треба виконати команду `git pull`, яка завантажить останні зміни з віддаленого репозиторію і смержить їх з вашими локальними правками, після чого можна повторити команду `git push`

# Як підключитись до потрібного репозиторію github?
<p id='part5'></p>

[stackoverflow](https://ru.stackoverflow.com/questions/1069192/%D0%9A%D0%B0%D0%BA-%D0%BF%D0%BE%D0%B4%D0%BA%D0%BB%D1%8E%D1%87%D0%B8%D1%82%D1%8C%D1%81%D1%8F-%D0%BA-%D0%BD%D1%83%D0%B6%D0%BD%D0%BE%D0%BC%D1%83-%D1%80%D0%B5%D0%BF%D0%BE%D0%B7%D0%B8%D1%82%D0%BE%D1%80%D0%B8%D1%8E-github)

Є репозиторій на гітхаб, я завантажив файли методом кнопки download замість того, щоб клонувати. І після цього файли ніяк не пов'язані з тим репозиторієм. Як зробити що б файли на моєму комп'ютері підключилися до того репозиторію? ( Зміна зробив у локальних файлах, тому клонувати той з гітхабу не варіант)

>Це робиться командою `git remote add`, але робити її можна тільки на існуючому репозиторії, так що вам знадобиться ще `git init`. Потім вам знадобиться отримати з репозиторію комміти командою `git fetch` і зробити `git reset`, щоб вносити свої зміни не з нуля, а починаючи з голови репозиторію.

`git init`<br>
`git remote add origin url_репозиторію`<br>
`git fetch origin`<br>
`git reset --mixed origin/master`<br>
`git add змінені файли`<br>
`git commit -m "коментар до комітету"`<br>
`git push -u origin master`<br>

URL репозиторію можна отримати, натиснувши на ту саму кнопку Clone в інтерфейсі github.

навіщо потрібні `git fetch origin` і `git reset --mixed origin/master` –
>без цих команд у вас буде порожня локальна копія, з якої не вийде нічого запушити (буде конфлікт)

# Як дізнатися, куди закомітилося?
<p id='part6'></p>

[stackoverflow](https://ru.stackoverflow.com/questions/605007/%D0%9A%D0%B0%D0%BA-%D1%83%D0%B7%D0%BD%D0%B0%D1%82%D1%8C-%D0%BA%D1%83%D0%B4%D0%B0-%D0%B7%D0%B0%D0%BA%D0%BE%D0%BC%D0%BC%D0%B8%D1%82%D0%B8%D0%BB%D0%BE%D1%81%D1%8C)

Поки не зробиш `git push`... воно не буде ні в якому обліковому записі, а буде просто локально на твоїй машині. Що додати два або більше репозиторію в проект потрібно скористатися командою `git remote add назва_репо(origin) url_адреса` і потім можна виконати команду `git push назва_репо(origin) назва_гілки`. Який обліковий запис доданий можна подивитися ось цією командою `git config --list` там висвітиться пошта та ім'я. Якщо хочеш пушити в два репозиторія відразу, то потрібно давати дозволи іншому акаунту в одному з активних акаунтів.

>`git remote -vv` - переглянути репозиторій.<br>
`git config --list` - переглянути пошту та ім'я користувача.<br>
`git remote add назва_репо(origin) url_адреса` - додати новий репозиторій.<br>
`git remote set-url назва_репо(origin) url_адреса` - змінити репозиторій.<br>

# Перенесення git-репозиторію

[stackoverflow](https://ru.stackoverflow.com/questions/571082/%D0%9F%D0%B5%D1%80%D0%B5%D0%BD%D0%BE%D1%81-git-%D1%80%D0%B5%D0%BF%D0%BE%D0%B7%D0%B8%D1%82%D0%BE%D1%80%D0%B8%D1%8F)

Є деякий репозиторій з гілками master та develop:`git@git.test.com:my-repo-a.git`<br>
Є також порожній репозиторій `git@git.test.com:my-repo-b.git`
    
Як повністю перенести репозиторій a до b?

Під цілком мається на увазі, що повинні зберегтися всі гілки та всі комміти.

>`git clone --bare git@git.test.com:my-repo-a.git`<br>
`git fetch origin`<br>
`git remote add new-origin git@git.test.com:my-repo-b.git`<br>
`git push --mirror new-origin`<br>
`git remote rm origin`<br>
`git remote rename new-origin origin`<br>
